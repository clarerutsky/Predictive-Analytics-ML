```python
# Step 1 : Import data

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree

drug_data=pd.read_csv('C:\\Users\\Dell\\Desktop\\drug200.csv')
drug_data.head()
```

```python
# Step 2 : Prepare the data 

# Prepare the independent dataset (input)

X=drug_data.drop(columns=['Drug'])
X.head()
```

```python
# Prepare the dependent dataset (output)

y=drug_data['Drug']
y.head()
```

```python
# Step 3 : Create the model

model=DecisionTreeClassifier()
```

```python
# Step 4 : Split the model into test and training data 

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Step 5 : Train our model

model.fit(X_train,y_train)
```

```python
# Step 6 : Make predictions

predictions=model.predict(X_test)
predictions
```

```python
# Step 7 : Evaluate the model

prediction_accuracy=accuracy_score(y_test,predictions)
prediction_accuracy
```

```python
# Visualize the model

plt.figure(figsize=(40,20))
dtree_plot = tree.plot_tree(model,
                           feature_names = ['Age','Sex','BP','Cholestrol','Na_to_K'],
                           class_names = sorted(y.unique()),
                           filled = True,
                           rounded = True,
                           fontsize = 14)

plt.savefig('C:\\Users\\Dell\\Desktop\\decision_tree_plot.png')
plt.show()
```
