```python

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('C:\\Users\\Kasutaja\\Downloads\drug200.csv')
data
```

```python
# Step 2 : Prepare the data 

# Prepare the dependent dataset (input)

y=data['Drug']
y
```

```python
# step 2 .Prepare the independent dataset (input)

x = data.drop(columns= ['Drug'])
x
```

```python
# Step 3 : Create the model

model=DecisionTreeClassifier()
```

```python
# Step 4 : Split the model into test and training data 
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Step 5 : Train our model

model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
# X_train data 

X_train
#y_train data 

y_train
# X_test data 
# y_test data 
y_test
X_test
```

```python
y_test
```

```python
# Step 6 : Make predictions

predictions=model.predict(X_test)
```

```python
# Predicted output
predictions
```

```python
# Step 7 : Evaluate the model

prediction_accuracy=accuracy_score(y_test,predictions)
prediction_accuracy
```
