### **Central Tendency**

Data in the world tends to clump around certain numbers - the average height of a man, or the average score on a test. 

This is called the central tendency.

##### **Mean**

This is the simple averaege of the data.

````python
# Getting the mean salary from this data
mean_salary = df['salary'].mean()
print('The mean salary is:',mean_salary)
````

##### **Median**

Median is an improved version of mean that discards the outliers and measures the central tendency by ordering the data.

````python
med_salary = df['salary'].median()
print ("The median salary is :", med_salary)
````

##### **Mode**

This measure of central tendency is calculated by the frequency. 

Mode is given by the maximum frequency of a field.

````python
# Let's take some hotel reviews

data = pd.DataFrame([
 { 'restaurant': 'MyHotel', 'reviewer': 'a', 'stars': 2 } ,
 { 'restaurant': 'MyHotel', 'reviewer': 'b', 'stars': 2 },
 { 'restaurant': 'MyHotel', 'reviewer': 'c', 'stars': 5 },
 { 'restaurant': 'MyHotel', 'reviewer': 'd', 'stars': 4 },
 { 'restaurant': 'MyHotel', 'reviewer': 'e', 'stars': 4 },
 { 'restaurant': 'MyHotel', 'reviewer': 'f', 'stars': 3 },
 { 'restaurant': 'MyHotel', 'reviewer': 'g', 'stars': 2 }    
])

data
````

````python
data['stars'].mode()
````

***You can see that 2 stars has the maximum frequency and hence you get a mode of 2 for the below data.***

#### **Measures of Spread**

````python
# Lets define two sample lists
list_one = pd.Series([-10, 0, 10, 20, 30])
list_two = pd.Series([8, 9, 10, 11, 12])
````

**Variance**

In probability theory and statistics, variance is the expectation of the squared deviation of a random variable from its mean. Informally, it measures how far a set of numbers is spread out from their average value.

<img src="https://www.gstatic.com/education/formulas/images_long_sheet/en/sample_variance.svg">

````python
# Finding the variance
print("Variance of list_one =", list_one.var()) 
print("Variance of list_two =", list_two.var())
````

**Standard Deviation**

In statistics, the standard deviation is a measure of the amount of variation or dispersion of a set of values.

<img src="https://www.gstatic.com/education/formulas/images_long_sheet/en/population_standard_deviation.svg">

````python
# Let's find the standard deviation (What is standard Deviation?)
import math
math.sqrt(list_one.var())
````

## **Basics of Data Visualization**

There are two libraries in python that are extensively used to create visualizations in python.

`matplotlib`
`seaborn`



**We are first going to see how to create simple data visualizations using Matplotlib.**

##### **Matplotlib**

````python
# Importing matplotlib
import matplotlib.pyplot as plt
````

````python
years = [2020, 2021, 2022, 2023, 2024]
total_population = [8939002, 8954511, 8960389, 8956742, 8943720]

plt.plot(years, total_population)
plt.title('Population over Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
````

````python
# Bar plots 

# Our data
labels = ["JavaScript", "Java", "Python", "R"]
usage = [62.8, 43.3, 36.8, 32.4]

# Generating the y positions. Later, we'll use them to replace them with labels.
y_positions = range(len(labels))

# Creating our bar plot
plt.bar(y_positions, usage) # Change bar to barh and see what happens?
plt.xticks(y_positions, labels)
plt.ylabel("Usage (%)")
plt.title("Usage of different programming languages")
plt.show()
````

If you want to show the correlation between two variables, you can show this using scatterplot.

````python
#Let's find the correlation between temperature and icecream sold.

temp = [30, 32, 33, 28.5, 35, 29, 29]
ice_creams = [100, 115, 115, 75, 123, 76, 82]

plt.scatter(temp, ice_creams)
plt.title("Correlation between temperature and ice cream sold")
plt.xlabel('temperature')
plt.ylabel('ice_creams')
plt.show()
````

To show the distribution of some data you can show it using histograms.

````python
numbers = [1,2,3,3.5,4,4.5,5,6.3,6.6,6.7,7,8,9,9.2,9.3]
plt.hist(numbers, bins=10)
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.show()
````

#### **Subplots using Matplotlib - When multiple plots need to be shown**

````python
# Importing required library 
import numpy as np 

x = np.linspace(0, 20, 100)

fig = plt.figure()  # create a plot figure

# Create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x)) # sine plot

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x)) # cosine plot
````

#### **Saving the figure to a file**

````python
fig.savefig('test_figure.png') # Give a name to the file
print('Figure has been saved')
````

#### **Seaborn**

Now lets venture into using the seaborn library to create the next set of visualizations!

We will try to do some exploratory data analysis using the Titanic dataset that we have seen before.

````python
#Importing the libraries
import pandas as pd
import seaborn as sns
import matplotlib.style as style

style.use('seaborn-poster')
style.use('ggplot')
````

##### **Data Understanding**

````python
# Let us load the data
url = "http://bit.ly/wkspdata"
data = pd.read_csv(url)
data.head()
````

````python
# Let us look into the data to see if there is anything missing
data.info() 
````

**This shows that there are a total of 891 records.**

`age, cabin and embarked have null values.`

````python
# Let's look into the descriptive statistics of the dataset
data.describe()
````

You can see that the average age of the passengers was 30 years.

The mean fare of the dataset is 32 whereas the median fare is 14.5.

Let's start analyzing the data visually!

````python
# Mapping data to strings to make visualizations more meaningful
data["alive"] = data.survived.map({0:"no", 1:"yes"}) # Adding a new column Alive to map the survived column
data["class"] = data.pclass.map({1:"First", 2:"Second", 3: "Third"}) # Adding a new column Class to map the pclass.
data.head()
````

##### **Factorplot**

````python
# Factorplot
sns.factorplot("sex", data = data, kind="count");
````

***We can see that the male passengers were double in number compared to the female passengers.***

````python
# Factorplot
sns.factorplot("class", data=data, hue="sex", kind="count");
````

***We can see that in Third class the number of male passengers are more than double as compared to the females.***


##### **KDE Plot**

````python
# KDE Plot

col = dict(male="#6495ED", female="#F08080")
fig = sns.FacetGrid(data, hue="sex", aspect=2, palette=col)
fig.map(sns.kdeplot, "age", shade=True)
fig.set(xlim=(0, 80));
fig.add_legend()
````

````python
fig = sns.FacetGrid(data, hue="class", aspect=2, palette="BuPu_d")
fig.map(sns.kdeplot, "age", shade=True)
fig.set(xlim=(0, 80));
fig.add_legend()
````

You can see from the above visuals that there are more male passengers travelling and passengers in their 20s are more in number.

We can draw an insight that passengers in their 20s who are solo travellers might have booked the third class tickets.

The box plot is a standardized way of displaying the distribution of data based on the five number summary: minimum, first quartile, median, third quartile, and maximum.

##### **Box-Plots**

````python
# Age by class (survived)
sns.factorplot(x="class", y="age", hue="survived", 
               data=data, aspect=2, kind="box")
````

##### **Correlation Analysis**

Correlation is a statistical technique that can show whether and how strongly pairs of variables are related. 

It is important to look at the correlation between different pieces of data to derive meaningful information.

````python
corr = data.corr() # Finding the correlation between different columns

#Generate Color Map, red & blue
colorsintheheatmap = sns.diverging_palette(220, 10, as_cmap=True)

#Generate Heat Map, allow annotations and place floats in map
sns.heatmap(corr, cmap=colorsintheheatmap, annot=True, fmt=".2f")
````

We can see a negative correlation between passengers survived and pclass. This implies that passengers travelling in first class had more survival rate. *Probably because lifeboats gave preference to the first class passengers more than the other passengers.*

##### **Violin Plots**

````python
import numpy as np
data["deck"] = data.cabin.str[0].map(lambda s: np.nan if s == "T" else s) # Getting a new column called deck
data.tail(5)
````

````python
# Let's see the distribution of passengers in the respective decks.
sns.violinplot(data["deck"], data["fare"], aspect=5)
````

````python
sns.violinplot(data["class"],data["fare"]).set_ylim(0, 600)
````

This shows that the frequency of passengers in the third class were the maximum.

##### **Scatterplot**

````python
# Relation between Fare and Age by the class, Survived.
sns.lmplot(x="fare", y="age", hue="survived", data=data)
````

````python
sns.lmplot("age", "survived", data, logistic=True, y_jitter=.05);
````
